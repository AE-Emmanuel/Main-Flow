```python
#importing necessary libraries 
from tkinter import *
from PIL import ImageTk, Image  

#function definition 
def init_calc():
    rt = Tk()
    rt.title("Calculator")

    #loading icon image 
    icon_img = PhotoImage(file='Documents\\jUPITER nOTEBOOK\\calculator_411744.png')
    rt.iconphoto(True, icon_img)
    rt.geometry('500x500')

    #background image 
    bg = ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\bg.jpg')
    bg_label = Label(rt, image=bg)
    bg_label.place(x=0, y=0)

    # Entry Widget
    e = Entry(rt, width=35, borderwidth=10, font=("Times", 10))
    e.grid(row=0, column=0, pady=20, padx=30, columnspan=3)

    #load images
    images = {
        "1": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-1-100.png'),
        "2": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-2-100.png'),
        "3": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-3-100.png'),
        "4": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-4-100.png'),
        "5": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-5-100.png'),
        "6": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-6-100.png'),
        "7": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-7-100.png'),
        "8": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-8-100.png'),
        "9": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-9-100.png'),
        "0": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-0-100.png'),
        "+": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-plus-100.png'),
        "-": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-minus-100.png'),
        "*": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-x-100.png'),
        "/": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-slash-100.png'),
        "=": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-parallel-symbol-100.png'),
        "Clear": ImageTk.PhotoImage(file='Documents\\jUPITER nOTEBOOK\\Numbers\\icons8-clear-100.png'),
    }

    def button_click(number):
        current = e.get()
        e.delete(0, END)
        e.insert(0, str(current) + str(number))

    def button_add():
        first_number = e.get()
        global f_num
        global maths
        maths = "addition"
        f_num = int(first_number)
        e.delete(0, END)

    def button_sub():
        first_number = e.get()
        global f_num
        global maths
        maths = "subtraction"
        f_num = int(first_number)
        e.delete(0, END)

    def button_div():
        first_number = e.get()
        global f_num
        global maths
        maths = "division"
        f_num = int(first_number)
        e.delete(0, END)

    def button_multi():
        first_number = e.get()
        global f_num
        global maths
        maths = "multiplication"
        f_num = int(first_number)
        e.delete(0, END)

    def button_equals():
        second_number = e.get()
        e.delete(0, END)
        
        if maths == 'addition':
            e.insert(0, f_num + int(second_number))
        if maths == 'subtraction':
            e.insert(0, f_num - int(second_number))
        if maths == 'multiplication':
            e.insert(0, f_num * int(second_number))
        if maths == 'division':
            e.insert(0, f_num / int(second_number))

    def button_clear():
        e.delete(0, END)

    #defining buttons 
    buttons = {
        "1": (1, 0, lambda: button_click(1)),
        "2": (1, 1, lambda: button_click(2)),
        "3": (1, 2, lambda: button_click(3)),
        "4": (2, 0, lambda: button_click(4)),
        "5": (2, 1, lambda: button_click(5)),
        "6": (2, 2, lambda: button_click(6)),
        "7": (3, 0, lambda: button_click(7)),
        "8": (3, 1, lambda: button_click(8)),
        "9": (3, 2, lambda: button_click(9)),
        "0": (4, 1, lambda: button_click(0)),
        "+": (4, 2, lambda: button_add()),  
        "-": (4, 0, lambda: button_sub()),  
        "*": (5, 1, lambda: button_multi()), 
        "/": (5, 0, lambda: button_div()),  
        "=": (5, 2, lambda: button_equals()),  
        "Clear": (6, 1, lambda: button_clear()),  
    }

    for btn, (row, col, cmd) in buttons.items():
        Button(rt, border='3', image=images[btn], command=cmd).grid(row=row, column=col)

    rt.images = images  # Keeps a reference to images to avoid garbage collection
    rt.mainloop()
    
init_calc()
```
